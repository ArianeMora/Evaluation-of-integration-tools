```python
import pandas as pd

cancer_genes = pd.read_csv('cBioPortalCancerGeneList.txt', header=None)
cancer_genes = cancer_genes[0].values
```

```python
# Read in the others
methyl_genes = pd.read_csv('MethylMix_PAAD_driver_genes.txt', sep='\t')
methyl_genes= methyl_genes['x'].values
```

```python
# Need to map the cancer_genes to the ensembl IDs
gene_map = pd.read_csv('tcga_hsapiens_gene_ensembl-GRCh38.p13.csv')
gene_map
```

```python
name_to_ens = {}
ensembl_gene_ids = gene_map['ensembl_gene_id']
for i, g in enumerate(gene_map['external_gene_name'].values):
    if g:
        name_to_ens[g] = ensembl_gene_ids[i]
        
cancer_ens = [name_to_ens.get(g) for g in cancer_genes]
cancer_ens
```

```python
len(set(set(sircle_genes) & set(cancer_ens)))
```

```python
len(set(set(methyl_genes) & set(cancer_ens)))
```
